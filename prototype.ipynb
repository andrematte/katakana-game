```python
import pandas as pd
```

```python
data_path = "data/katakana-words.csv"
first_n_words = 3000
n_levels = 6
labels = [1,2,3,4,5,6]

df = pd.read_csv(data_path, nrows=first_n_words)
df['Level'] = pd.qcut(df['Frequency'], n_levels, labels=labels)
df.head()
```

```python
cards = pd.DataFrame()

for level in labels:
    words = df[df['Level'] == level]
    words = words.sample(frac=1).reset_index()
    
    cards[f'Level {level}'] = words['Word']
    cards = cards.dropna()
    
cards
```
